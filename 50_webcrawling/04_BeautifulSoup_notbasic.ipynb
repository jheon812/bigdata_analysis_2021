# 4. BeautifulSoup 활용
* 멜론사이트 크롤링하기
* 제목, 가수, 앨범, 좋아요수 추출

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/index.htm'

headers = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'
}
res = requests.get(url,headers=headers)
res
```

```python
res.text[:300] #\r\n이 (타자기 왼쪽으로 -> 다음줄) 한 번에 엔터 / 리눅스에서는 \r만 윈도우즈에서 \r\n
```

```python
soup = BeautifulSoup(res.text)
```

## 노래제목 추출

```python
songs = soup.select('.ellipsis.rank01') #select 함수에서는 space를 점으로 바꾸기
title = []
for song in songs :
    title.append(song.find('a').get_text())
title
```

## 가수명 추출

```python
songs = soup.select('.ellipsis.rank02') #select 함수에서는 space를 점으로 바꾸기
singer = []
for song in songs :
    singer.append(song.find('a').get_text())
singer
```

## 앨범명 추출

```python
songs = soup.select('.ellipsis.rank03') #select 함수에서는 space를 점으로 바꾸기
album = []
for song in songs :
    album.append(song.find('a').get_text())
album
```

## 좋아요 수 추출

```python
songs = soup.select('.button_etc.like') #select 함수에서는 space를 점으로 바꾸기
params = '' # AJAX

for song in songs :
    params += song['data-song-no'] + ','
params[:-1]
```

```python
url2 = 'https://www.melon.com/commonlike/getSongLike.json?contsIds=' + params[:-1]

res = requests.get(url2,headers=headers)
res
```

```python
songs = res.json()['contsLike']
like = []

for song in songs :
    like.append(song['SUMMCNT'])
like
```

```python
#제목, 가수, 앨범, 좋아요 dataframe 생성

df_data = {
    'title':title,
    'singer': singer,
    'album': album,
    'like':like
}

df = pd.DataFrame(data=df_data)
df
```
